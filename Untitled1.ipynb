```python
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)

loo = LeaveOneOut()

ridge_model = Ridge(alpha=0.1)

ridge_scores = cross_val_score(ridge_model, scaled_X_train, y_train, scoring='neg_mean_squared_error', cv=loo)
ridge_rmse = np.sqrt(-ridge_scores.mean())

# Cross-validation for R²
ridge_r2_scores = cross_val_score(ridge_model, scaled_X_train, y_train, scoring='r2', cv=loo)
ridge_r2_mean = ridge_r2_scores.mean()

# Fit the model on the entire training set
ridge_model.fit(scaled_X_train, y_train)

# Predict on the test set
y_pred = ridge_model.predict(scaled_X_test)

# Calculate RMSE and R² on the test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
test_r2 = r2_score(y_test, y_pred)

print(f"Cross-Validation RMSE: {ridge_rmse}")
print(f"Cross-Validation R²: {ridge_r2_mean}")
print(f"Test RMSE: {test_rmse}")
print(f"Test R²: {test_r2}")
```
